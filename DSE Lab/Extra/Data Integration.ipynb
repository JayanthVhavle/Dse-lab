```python
df2=pd.read_csv(file_2,header=0)
```

```python
df2.shape
```

```python
df2.columns
```

```python
df2.head()
```

```python
#extracting only required attibutes as per requirements
col2=['movie_name','wordsInTitle','url','type','imdbRating','ratingCount', 'nrOfWins',
       'nrOfNominations','nrOfUserReviews','duration','Action', 'Adult', 'Comedy', 'Sport','War','Drama', 'Family','Horror','RealityTV','Thriller']
```

```python
df2=df2[col2]
```

```python
#removing inconsistency in data

col3=['Action', 'Adult', 'Comedy', 'Sport','War','Drama', 'Family','Horror','RealityTV','Thriller']

for i in col3:
    df2[i]=df2[i].apply(lambda t: 1 if t=="yes" else 0)
```

```python
df2.head()
```

```python
df2.shape
```

```python
df2.rename(columns={'movie_name':'title'},inplace=True)
```

```python
df2
```

```python
df=pd.merge(df1,df2,on='title')
```

## Checking the type of the data values and changing them to consistent format

```python
for col in df.columns:
    print(col,"  ",type(df[col][5]))
```

```python
df['imdbRating']=pd.to_numeric(df['imdbRating'],errors='coerce')
type(df['imdbRating'][5])
```

```python
df['ratingCount']=pd.to_numeric(df['ratingCount'],errors='coerce')
type(df['ratingCount'][5])
```

```python
df['nrOfWins']=pd.to_numeric(df['nrOfWins'],errors='coerce')
type(df['nrOfWins'][5])
```

```python
df['nrOfNominations']=pd.to_numeric(df['nrOfNominations'],errors='coerce')
type(df['nrOfNominations'][5])
```

```python
df['duration']=pd.to_numeric(df['duration'],errors='coerce')
type(df['duration'][5])
```

```python
for col in df.columns:
    print(col,"  ",type(df[col][5]))
```

## Handling the missing values

```python
for col in df.columns:
    print(col,"  ",df[col].isna().sum())
```

```python
df['imdbRating']
```

```python
df['imdbRating'].fillna(df['imdbRating'].mean(),inplace=True)
```

```python
df['imdbRating']
```

```python
df['ratingCount'].fillna(df['ratingCount'].mean(),inplace=True)
```

```python
df['duration'].fillna(df['duration'].median(),inplace=True)
```

```python
df_new=df[['lifetime_gross','year','imdbRating','ratingCount','nrOfWins','nrOfNominations','nrOfUserReviews','duration']]
```

```python
df['award_performance']=df.apply(lambda row:row['nrOfWins']/row['nrOfNominations'] if row['nrOfNominations']>0 else 0,axis=1)
```
